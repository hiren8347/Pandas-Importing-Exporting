```python
from sqlalchemy import create_engine
import pandas as pd
import pymysql
connection=pymysql.connect(host='localhost',
                          user='root',
                          password='root',
                          db='register',
                          cursorclass=pymysql.cursors.DictCursor)
data=pd.read_sql_query('select * from registermaster',connection)
print(data)
engine=create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                    .format(user="root",
                          pw="root",
                          db="pythondb"))


data.to_sql(con = engine , name='registermaster', if_exists ='replace')
```
